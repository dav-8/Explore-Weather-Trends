```python
# importing packages
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
sns.set() # set seaborn as default because sea born plots more fansy than matplotlip
```

```python
#impoting our data 
df1 = pd.read_csv('Global.csv')
df2= pd.read_csv('Cairo.csv')
```

```python
# make copy to keep the original data 
global_df = df1.copy()
egypt_df=df2.copy()
```

```python
global_df.head()
```

```python
egypt_df.head()
```

```python
#checking for nulls
global_df.isnull().sum()
```

```python
egypt_df.isnull().sum()
```

```python
# creating new col for moving average for 7 days
global_df['moving_avg']=global_df.avg_temp.rolling(7).mean()
```

```python
pd.set_option('display.max_rows', None) # this code help be to see all rows
global_df.moving_avg
```

```python
egypt_df['moving_avg']=egypt_df.avg_temp.rolling(7).mean()
```

```python
egypt_df.moving_avg
```

```python
plt.figure(figsize=(10,5))
plt.plot(global_df.year,global_df.moving_avg,label ="global_MA")
plt.plot(egypt_df.year,egypt_df.moving_avg,label ="Cairo_MA")
plt.title('Cairo temp vs. global temp')
plt.xlabel('Years')
plt.ylabel('Moving avg temp')
plt.xlim(egypt_df.year.min())
plt.legend()
plt.show()
```
